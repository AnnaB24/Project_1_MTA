```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# merge data sets for Tableau: demographic (Census), MTA locations(mta.info), top 20 stations (MTA Data Project Notebook_new.ipynb)

top20_stations=pd.read_csv('top20_stations.csv')
percent_female = pd.read_csv('NY_census_data.csv', skiprows=1)
percent_female.info()
```

```python
percent_female.head()
```

```python
percent_female2 = percent_female[['Id2','Percent; SEX AND AGE - Female population']]
```

```python
percent_female_zip = percent_female2.rename(columns={'Id2':'ZIP'})
```

```python
percent_female_zip.head(3)
```

```python
mta_loc = pd.read_csv('http://web.mta.info/developers/data/nyct/subway/Stations.csv')
mta_loc.head(3)
```

```python
mta_loc['Stop Name'] = mta_loc['Stop Name'].str.replace(' - ','-')
mta_loc['Stop Name'] = mta_loc['Stop Name'].str.upper() 
mta_loc.head()
mta_loc.rename(columns={'Stop Name': 'STATION'}, inplace = True)
mta_loc.STATION = mta_loc.STATION.replace('34 ST-PENN STATION', '34 ST-PENN STA')
mta_loc.STATION = mta_loc.STATION.replace('GRAND CENTRAL-42 ST', 'GRD CNTRL-42 ST')
mta_loc.STATION = mta_loc.STATION.replace('42 ST-PORT AUTHORITY BUS TERMINAL', '42 ST-PORT AUTH')
mta_loc.STATION = mta_loc.STATION.replace('59 ST-COLUMBUS CIRCLE', '59 ST COLUMBUS')
mta_loc.STATION = mta_loc.STATION.replace('47-50 STS-ROCKEFELLER CTR', '47-50 STS ROCK')
mta_loc.STATION = mta_loc.STATION.replace('FLUSHING-MAIN ST', 'FLUSHING-MAIN')
mta_loc.STATION = mta_loc.STATION.replace('JACKSON HTS-ROOSEVELT AV', 'JKSN HT-ROOSVLT')
mta_loc.head(3)
```

```python
mta_loc = mta_loc[['STATION','GTFS Latitude', 'GTFS Longitude']]
mta_loc.drop_duplicates(subset=['STATION'], inplace=True)
mta_loc.head(3)
```

```python
top20station_loc = pd.merge(top20_stations, mta_loc, on='STATION',  how='inner')
```

```python
top20station_loc.head(3)
```

```python
stations_zip= pd.read_csv('stations_ZIP.csv')
stations_zip.head(3)
```

```python
percent_female_station = pd.merge(stations_zip, percent_female_zip, on='ZIP')
```

```python
top20station_entery_loc_zip_fem = pd.merge(top20station_loc, percent_female_station, on='STATION',  how='left')
```

```python
# result for Tableau
top20station_entery_loc_zip_fem.head(3)
```

```python
top20station_entery_loc_zip_fem.to_csv("top20station_entery_loc_zip_fem.csv")
```
